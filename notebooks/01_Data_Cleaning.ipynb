```python
import pandas as pd

# 讀取檔案
df = pd.read_csv('YRBS_2007.csv')

# 顯示前五筆資料，確認是否讀取成功
df.head()
```

```python
# 篩選我們需要的欄位
target_columns = ['BMIPCT', 'WhatIsYourSex', 'PhysicalActivity5OrMoreDays', 'TelevisionWatching']
df_selected = df[target_columns].copy()

# 檢查各欄位的資料型態與缺失值數量
print("--- 原始欄位資訊與缺失值統計 ---")
print(df_selected.info())
print("\n各欄位初始缺失值數量：")
print(df_selected.isnull().sum())
```

```python
# 1. 檢查並清理依變數 BMIPCT
# 確保 BMI 百分位數在正常 0~100 範圍內，排除不合理的極端值
df_selected = df_selected[(df_selected['BMIPCT'] >= 0) & (df_selected['BMIPCT'] <= 100)]

# 2. 檢查性別編碼是否正常（應只有 1 和 2）
df_selected = df_selected[df_selected['WhatIsYourSex'].isin([1, 2])]

# 3. 移除其他自變數中的 NaN (缺失值)
df_cleaned = df_selected.dropna()

print(f"\n清洗完成後的有效資料筆數：{df_cleaned.shape[0]} 筆")
```

```python
import os

# 確保資料夾存在
os.makedirs('data/processed', exist_ok=True) 

# 將 df_final 改為 df_cleaned 進行儲存
df_cleaned.to_csv('data/processed/yrbs_cleaned.csv', index=False)

print("\n已成功儲存至 data/processed/yrbs_cleaned.csv")
```
